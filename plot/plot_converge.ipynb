```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
filenames = ["WI_4W_M3-1_estimate.txt", "BI_1D_M4-4_estimate.txt", "BI_1W_M3-1_estimate.txt", "BI_1W_M4-4_estimate.txt"]
cnts = []
for i in range(len(filenames)):
    sample_cnts = []
    with open("./converge/"+filenames[i], 'r') as f:
        lines = f.readlines()
        for line in lines:
            cnt = list(map(float,line.strip('\n').split()))
            sample_cnts.append(cnt[0])
    cnts.append(sample_cnts)
cnts = np.array(cnts) 
```

```python
exact_cnts = [639782710513, 22645832525127]
num_samples = [i*10 for i in range(1, 101)]
for i in range(2):
    fig, ax = plt.subplots(1, 1, figsize=(4, 3))
    # axs[j].axhline(y=exact_cnts[j+1], color='grey', linestyle='--')
    ax.axhline(y=0, color='grey', linestyle='--')
    ax.scatter(num_samples, (cnts[i]/exact_cnts[i]-1)*100, marker='.',color='blue')
    ax.set_xlabel("Num of samples (M)", fontsize=18)
    ax.set_ylabel("Relative error (%)", fontsize=18)
    # ax.set_ylim(bottom=-10, top=10)
    ax.tick_params(axis='x', labelsize=16)
    ax.tick_params(axis='y', labelsize=16)
    ax.xaxis.offsetText.set_fontsize(12)

    plt.tight_layout()
    plt.savefig(f"figs/converge_{i}.pdf")
```

```python
num_samples = [i*10 for i in range(1, 101)]
for i in range(2, 4):
    fig, ax = plt.subplots(1, 1, figsize=(4, 3))
    # axs[j].axhline(y=exact_cnts[j+1], color='grey', linestyle='--')
    ax.axhline(y=0, color='grey', linestyle='--')
    ax.scatter(num_samples, cnts[i], marker='.',color='blue')
    ax.set_xlabel("Num of samples (M)", fontsize=18)
    ax.set_ylabel("Estimated counts", fontsize=18)
    ax.set_ylim(bottom=max(cnts[i])*0.5)
    ax.tick_params(axis='x', labelsize=16)
    ax.tick_params(axis='y', labelsize=16)
    ax.xaxis.offsetText.set_fontsize(12)

    plt.tight_layout()
    plt.savefig(f"figs/converge_{i}.pdf")
```
